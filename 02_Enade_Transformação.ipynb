# Transformação

```python
#Criando uma nova variável substituindo os códigos pelos Privado
enade['DESC_PUBLICA'] = ""
enade.loc[enade.CO_CATEGAD.isin([118,120,121,10005,10006,10007,10008,10009, 17634]),"DESC_PUBLICA"] = "Privado"
```

```python
#Criando uma nova variável substituindo os códigos pelos Público
enade.loc[
    enade.CO_CATEGAD.isin([93,115,116,10001,10002,10003]), "DESC_PUBLICA"] = "Publica"
```

```python
enade.DESC_PUBLICA.value_counts()
```

```python
#Modificamos os atributos da variável CO_MODALIDADE com replace e atribuímos a variável CO_MODALIDADE
#enade['CO_MODALIDADE'] A COLCHETE É UMA FORMA DE ACESSARMOS A COLUNA
enade['CO_MODALIDADE'] = enade.CO_MODALIDADE.replace({
    0 : 'EaD',
    1 : 'Presencial'
})
```

```python
enade.CO_MODALIDADE.value_counts()
```

```python
enade['CO_REGIAO_CURSO'] = enade.CO_REGIAO_CURSO.replace({
    1 : 'Norte',
    2 : 'Nordeste',
    3 : 'Sudeste',
    4 : 'Sul',
    5 : 'Centro-Oeste'
})
```

```python
enade.CO_REGIAO_CURSO.value_counts() / enade.shape[0]
```

```python
#pd NA atribui valor Missing
enade['DESC_COR'] = enade.QE_I02.replace({
    "A" : 'Branca',
    "B" : 'Preta',
    "C" : 'Amarela',
    "D" : 'Parda',
    "E" : 'índigena',
    "F" : pd.NA,
    " " : pd.NA
})
```

```python
enade.DESC_COR.value_counts()
```
